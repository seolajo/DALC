# 26번 원형 데크 디자인

* MyCircularDeque(k): 데크 사이즈를 k로 지정하는 생성자  

* insertFront(): 데크 처음에 아이템을 추가하고 성공할 경우 true를 리턴
* insertLast(): 데크 마지막에 아이템을 추가하고 성공할 경우 true를 리턴
* deleteFront(): 데크 처음에 아이템을 삭제하고 성공할 경우 true를 리턴
* deleteLast(): 데크 마지막에 아이템을 삭제하고 성공할 경우 true를 리턴
* getFront(): 데크 첫 번째 아이템을 가져온다. 데크가 비었다면 -1 리턴
* getRear(): 데크 마지막 아이템을 가져온다. 데크가 비었다면 -1 리턴
* isEmpty(): 데크가 비었는지 여부 판별
* isFull(): 데크가 가득 차 있는지 여부 판별


```python
#리스트로 원형데크 구현하기
from __future__ import annotations
from typing import Any, Type

class MyCircularDeque:
    
    #생성자
    def __init__(self, k: int):
        self.k = k          # 데크 사이즈
        self.deque = []        # 데크 자체
        
    #데크의 처음에 아이템 추가
    def insertFront(self, value: int) -> bool:
        if self.isFull():                 #데크가 찼는지 확인하는 메소드. 하단에 구현.
            return False
        else:
            self.deque.insert(0, value)   #list.insert(인덱스, 추가할 값) : 해당 인덱스에 값이 추가되고, 그 뒤의 값들은 인덱스에 1 씩 더해진다.
        return True
        
    #데크의 마지막에 아이템 추가
    def insertLast(self, value: int) -> bool:
        if self.isFull():
            return False
        else:
            self.deque.append(value)      #list.append(추가할 값)  :  해당 리스트의 마지막에 값이 추가된다.
        return True
        
    #데크의 첫 번째 값을 삭제
    def deleteFront(self) -> bool:
        if self.isEmpty():                #데크가 비었는지 확인하는 메소드. 하단에 구현.
            return False
        else:
            self.deque.pop(0)             #list.pop(인덱스) 입력한 인덱스의 값을 삭제하면서 리턴하여 보여준다.
        return True
        
    #데크의 마지막 값을 삭제
    def deleteLast(self) -> bool:
        if self.isEmpty():                #데크가 비었는지 확인하는 메소드. 하단에 구현.
            return False
        else:
            self.deque.pop()              #list.pop(인덱스) 입력한 인덱스의 값을 삭제하면서 리턴하여 보여준다.
        return True
        
    #데크의 처음 값을 리턴
    def getFront(self) -> int:
        if self.isEmpty():                #데크가 비었는지 확인하는 메소드. 하단에 구현.
          return -1
        return self.deque[0]              #list[0] 리스트의 맨 처음 값 리턴
    
    #데크의 마지막 값을 리턴
    def getRear(self) -> int:

        if self.isEmpty():                #데크가 비었는지 확인하는 메소드. 하단에 구현.
          return -1
        return self.deque[-1]             #list[-1] 리스트의 맨 처음 값 리턴
    
    #데크가 비었으면 True 반환
    def isEmpty(self) -> bool:
        return (len(self.deque) == 0)     #list 길이가 0이면 True
        
    #데크가 가득 찼으면 True 반환
    def isFull(self) -> bool:
        return (len(self.deque) == self.k)   #list 길이가 데크의 크기 k와 일치하면 True

    #디버깅용 프린트메소드 추가
    def printDeque(self) -> None:
        if self.isEmpty():
            print('데크가 비었습니다.')
        else:
            for i in range(len(self.deque)):
                print(f'({i+1})', self.deque[i], end=' ')
            print()
```

#27번 k개 정렬 리스트 병합

k개의 정렬된 리스트를 1개의 정렬된 리스트로 병합하라.

* heapq.heappush(heap, item) : 힙 불변성을 유지하면서, item 값을 heap으로 푸시  

* heapq.heappush(heap, (i, item)) : i에 우선순위, 즉 인덱스를 입력하고, 내용(item)을 입력하면 우선순위대로 힙에 저장  

* heapq.heappop(heap) : 힙의 가장 작은 항목(heap[0])을 삭제하고 반환

```python
# Definition for singly-linked list.
from __future__ import annotations
from typing import Any, Type
import heapq


class ListNode:
    def __init__(self, val=0, next: ListNode = None):#(노드구성)
        self.val = val                               #값
        self.next = next                             #포인터

class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        head = ListNode(None)                        #헤드노드 생성
        curr = head                                  #주목노드(가리킬 노드) 생성 초기화
        h = []                                       #힙 생성
        
        #각 연결 리스트의 루트를 힙에 저장
        for i in range(len(lists)):
            if lists[i]:                             #리스트가 비지 않으면 실행
                heapq.heappush(h, (lists[i].val, i)) #값을 푸시(힙, 우선순위, 값)
                lists[i] = lists[i].next             #다음 노드 주목(바로 아래 while문과 이어짐)
       
       #디버깅용 프린트
        print(h,"\n")

       #힙 추출 이후 다음 노드는 다시 저장
        while h:                                     #힙이 빌 때까지 실행
            val, i = heapq.heappop(h)                #힙 추출 : 힙의 첫 번째 lists[i].val과 i를 val과 i에 저장, 및 힙에서 노드 삭제
            
            curr.next = ListNode(val)                #curr의 포인터에 값이 val(lists[i].val)인 노드 대입
            curr = curr.next                         #노드 업데이트
            
            if lists[i]:                             # 위의 for문에서 lists[i].next라고 했으므로 두번째 노드부터 실행
                heapq.heappush(h, (lists[i].val, i)) #힙에 다음 노드는 다시 저장
                lists[i] = lists[i].next             #다음 노드 주목
            #디버깅용 프린트
            print(h)

        return head.next

    def makeList(self, ptr: ListNode) -> Sequence: # 모든 노드 출력    
        makelist = []
        while ptr is not None:
            makelist += [ptr.val]
            ptr = ptr.next
        return makelist
```

```python
#솔루션 객체 생성
a = Solution()
```

```python
#디버깅용 리스트노드 생성

#listNode[0]
l1 = ListNode(1)
l1.next = ListNode(3)
l1.next.next = ListNode(5)

#listNode[1]
l2 = ListNode(2)
l2.next = ListNode(4)
l2.next.next = ListNode(6)

#listNode[2]
l3 = ListNode(10)
l3.next = ListNode(20);

#리스트노드형 리스트 생성
listNode = [l1, l2, l3]
```

```python
#mergeKLists에 listNode 대입
nodelist = a.mergeKLists(listNode)
#리스트노드 출럭
a.makeList(nodelist)
```
