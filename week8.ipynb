# 31번 상위 K 빈도 요소

```python
from __future__ import annotations
from typing import Any, Type
import collections

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        return list(zip(*collections.Counter(nums).most_common(k)))[0] #빈도수 상위 k개 추출
```

```python
a = Solution()
```

```python
numlist = [1,1,1,2,2,3]
```

```python
a.topKFrequent(numlist, 2)
```
