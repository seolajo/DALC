# Arrays 101 > Check If N and Its Double Exist




## 문제 설명

Given an array `arr` of integers, check if there exists two integers `N` and `M` such that `N` is the double of `M` ( i.e. `N = 2 * M`).

More formally check if there exists two indices i and j such that :

* `i != j`
* `0 <= i, j < arr.length`
* `arr[i] == 2 * arr[j]`
<br/><br/>


## Example

```
Input: arr = [10,2,5,3]
Output: true
Explanation: N = 10 is the double of M = 5,that is, 10 = 2 * 5.
```

<br/><br/>

## Constraints


* `2 <= arr.length <= 500`
* `-10^3 <= arr[i] <= 10^3`

## 정답 코드

````python
from typing import List
````

````python
class Solution:
    def checkIfExist(self, arr: List[int]) -> bool:
        arr = [i for i in arr if i*2 in arr]
        if arr:
            if arr != [0]:
                return True
        return False
````

````python
a = Solution()
````

````python
a.checkIfExist([-2,0,10,-19,4,6,-8])
````

## 다른 사람의 풀이

````python
class Solution:
    
    def checkIfExist(self, arr: List[int]) -> bool:
        
        s = set()
        
        for x in arr:
            
            if x * 2 in s or x / 2 in s:
                return True
            s.add(x)
            
        return False
````

````python
b = Solution()
````

````python
b.checkIfExist([-2,0,10,-19,4,6,-8])
````
