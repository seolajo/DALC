#20번 유효한 괄호

```python
from typing import *

class Solution:
  def isValid(self, s:str) -> bool:
    new = ''
    while new != s:
        new = s
        s = s.replace('()', '')
        s = s.replace('[]', '')
        s = s.replace('{}', '')
        if s == '':
            return True
    return False


a = Solution()

s = input('괄호를 입력하시오: ')
print(a.isValid(s))
```

# 21번 중복 문자 제거

```python
from typing import *

class Solution:
  def removeDuplicateLetters(self, s: str) -> str:
    new = []
    if s:
      new = list(s)
      new = set(new)
      new = sorted(new)
      answer = ''.join(new)
      return answer
    else:
      return False

a = Solution()

s = input('문자열을 입력하시오: ')
print(a.removeDuplicateLetters(s))
```

# 22번 일일 온도

```python
from typing import *

class Solution:
  def dailyTemperatures(self, T: List[int]) -> List[int]:
    answer = []

    for i in range(len(T)):
      for j in range(i, len(T)):
        if T[j] > T[i]:
          answer.append(j-i)
          break
        elif j == len(T) - 1:
          answer.append(0)

    return answer

ins = Solution()
T = []

while True:
  put = input('매일의 화씨 온도를 입력하시오(정수로 입력/입력완료...N): ')
  if put.isdecimal():
    T.append(int(put))
  elif put == 'N' or 'n':
    break
  else:
    print("잘못된 입력입니다.")
print(T)
print(ins.dailyTemperatures(T))
```
