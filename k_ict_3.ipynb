```python
from seaborn import load_dataset
iris = load_dataset("iris")
iris.info()
```

```python
iris.head()
```

```python
#Ccategorical함수를 이용해 columns를 범주형 데이터로 변환
import pandas as pd

iris["species_cat"] = pd.Categorical(iris.species)
iris.info()
#species는 문자열이라 object
#categoty는 R의 factor과 비슷한 인코딩된 타입이라 보면 됨
```

```python
iris.head()
```

```python
#cat 객체는 category>cat>codes라는 수치형 데이터 가지기에 c에 값을 지정할 수 있음 --> iris.species_cat.cat.codes
#cat 객체는 품종인데, cat객체를 수치데이터로 넣으면, 품종에 따라 색이 정해짐, 경향에 대한 직관을 얻을 수 있음
plt.scatter(iris.sepal_length, iris.sepal_width, c = iris.species_cat.cat.codes, alpha=0.2, cmap = 'viridis')
plt.title("iris")
plt.xlim(iris.sepal_length.min(), iris.sepal_length.max())
plt.ylim(iris.sepal_width.min(), iris.sepal_width.max())
plt.xlabel(iris.columns[0])
plt.ylabel(iris.columns[1])
```

```python
#11122344555가 있을 때 unique를 쓰면 12345만 남는다
species = iris.species.unique()

for sp in species:
    x = iris.loc[iris["species"] == sp, "sepal_length"]
    y = iris.loc[iris["species"] == sp, "sepal_width"]
    plt.scatter(x, y, label = sp, alpha=0.2)

plt.title("iris")
plt.xlim(iris.sepal_length.min(), iris.sepal_length.max())
plt.ylim(iris.sepal_width.min(), iris.sepal_width.max())
plt.xlabel(iris.columns[0])
plt.ylabel(iris.columns[1])
plt.legend()    
```

```python
#random.randn(n): standard normal distribution. 표준정규분포. 데이터가 n개인 표준정규분포.
#random.normal(i, j, n): normal distribution. 정규분포. 평균이 i, 표준편차가 j 인, 데이터가 n개인 정규분포. i=0, j=1로 설정하면 표준정규분포가 된다.

mu = 72
sigma = 8
x = np.random.normal(mu, sigma, 10000)
plt.hist(x, rwidth=0.9, bins=16)
#x : 만들어 놓은 랜덤값
#rwidth : 히스토그램 전체 크기(가로 길이)를 1이라 할 때 어느 정도의 크기로 설정할지. 1일때 각 막대기간 간격은 0. 크기가 작아질수록 간격커지고 막대기 얇아짐
#bins : 전체 계급구간의 개수

#시각화시 보여지는 y축 : 빈도수
```

```python
#tips 데이터셋 : 특징별 팁을 낸 빈도와 금액크기
tips = load_dataset("tips")
tips.info()
```

```python
#상대도수 히스토그램
plt.hist(x = tips.tip, rwidth=0.9, bins = 20)
plt.title('tips')
plt.xlabel('tip')
plt.ylabel('frequency')
```

```python
#확률밀도 함수 표현하기
import scipy as sp

bins = 20
plt.hist(x=tips.tip, rwidth=0.9, bins = bins, density=True)
plt.title('tips')
plt.xlabel('tip')
plt.ylabel('density')   #확률밀도함수를 구하려면 평균, 표준편차 알아야함으로 아래서 구함

mu, sigma = tips.tip.mean(), tips.tip.std() #mean : 평균구함, std(standard deviation) : 표준편차
tip_min = tips.tip.min()
tip_max = tips.tip.max()
x = np.linspace(tip_min, tip_max, bins) #x의 경계구간 구하기
y = sp.stats.norm(mu, sigma).pdf(x)     #scipy의 normal distribution의 확률밀도함수(pdf(x)) 활용해 y값 얻어(난수생성해서 구하는 정규분포는 numpy활용했음) 라인플룻 그리기
plt.plot(x, y)
```

```python
labels = []
tip_list=[]

for label, df_per_sex in tips.groupby("sex"):#성별로 묶으면, 열은 total_bill, tip, size 세 개임
    labels.append(label)          #female과 male.
    tip_list.append(df_per_sex["tip"].tolist()) #df_per_sex["tip"] : tip정보 담김. tolist() : 동등한 레벨의 정보를 묶어서 리스트 객체로 줌

plt.boxplot(tip_list, labels=labels)
plt.title("tips")
plt.xlabel("sex")
plt.ylabel("tip")
```
