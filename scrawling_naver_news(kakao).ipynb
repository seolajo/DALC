```python
import requests
from bs4 import BeautifulSoup

#num : 크롤링할 페이지수

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}

news = []
num = 254    # 마지막 페이지가 253페이지

for i in range(1, num):

    data = requests.get(f'https://finance.naver.com/item/news_news.nhn?code=035720&page={i}&sm=title_entity_id.basic&clusterId=', headers=headers)

    soup = BeautifulSoup(data.text, 'html.parser')

    trs = soup.select('body > div > table.type5 > tbody > tr')

    for tr in trs:
        title = tr.select_one('td.title > a').text
        dt = tr.select_one('td.date').text.split()

        date = dt[0]
        time = dt[1]

        a = tr.select_one('td.title > a')

        # if a is not None:
        base_url = 'https://finance.naver.com'
        url = base_url + a['href']
        data = requests.get(url, headers=headers)

        soup = BeautifulSoup(data.text, 'html.parser')
        content = soup.select_one('#news_read').text
        new = [title, date, time, content]
        news.append(new)
```

```python
import pandas as pd
import numpy as np
```

```python
data = np.array(news)
df = pd.DataFrame(data, columns=['제목', '날짜', '시간', '내용'])
```

```python
# df.to_csv('/content/drive/My Drive/Colab Notebooks/news.csv', sep='\t', encoding='utf-8')
writer = pd.ExcelWriter('/content/drive/My Drive/Colab Notebooks/news.xlsx')
df.to_excel(writer, 'DataFrame')
writer.save()
```
