# 28번 해시맵 디자인

```python
from __future__ import annotations
from typing import Any, Type

class MyHashMap:
  def __init__(self, size: int):
    self.size = size
    self.table = [None]*size

  def put(self, key: int, value: int) -> None:
    self.table[key] = value
    print(f'{value}를 삽입했습니다.')
  
  def get(self, key: int) -> int:
    if self.table[key] == None:
      return -1
    else:
      return self.table[key]

  def remove(self, key: int) -> None:
    if self.get(key) == -1:
      print('해당하는 키가 존재하지 않습니다.')
    else:
      self.table[key] = None
      print('삭제되었습니다.')
  
```

```python
hashMap = MyHashMap(100)
```

```python
hashMap.put(1, 1)
```

```python
hashMap.put(2, 2)
```

```python
hashMap.get(1)
```

```python
hashMap.get(3)
```

```python
hashMap.get(2)
```

```python
hashMap.remove(2)
```

# 29번 보석과 돌

```python
class JewelandStone:

  def numJewelInStones(self, J: str, S: str) -> int:
    count = 0
    for char in S:
      if char in J:
        count += 1
    
    return count
```

```python
test = JewelandStone()
```

```python
test.numJewelInStones('aA','aAAbbbb')
```

# 30번 중복 문자 없는 가장 긴 부분 문자열

```python
class getSubstring:

  def lengthOfLongstSubstring(self, s:str) -> int:
    max = 0
    stringlist = []
    for i in range(len(s)):
      for j in range(i,len(s)):
        if s[j] not in stringlist:
          stringlist += [s[j]]
        else:
          if len(stringlist) > max:
            max = len(stringlist)
          stringlist = []
          break

    return max
```

```python
test = getSubstring()
```

```python
test.lengthOfLongstSubstring('pwwkew')
```
