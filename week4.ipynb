13번 팰린드롬 연결 리스트

```python
from typing import Sequence

def isPalindrome(a: Sequence) -> bool:
  
  new = []
  for i in a:
    if i.isdigit():
      new.append(i)

  renew = list(reversed(new))

  return new == renew

if __name__ == '__main__':
  numlist = input('정렬을 입력하세요.: ')

  print(isPalindrome(numlist))
```

14번 두 정렬 리스트의 병합

```python
from typing import Sequence

def isPalindrome(a:Sequence, b: Sequence) -> str:
  
  new = []
  for i in (a+b):
    if i.isdigit():
      new.append(i)

  new.sort()
  new = '->'.join(new)    
  
  return new

if __name__ == '__main__':
  num1 = input('첫 번째 정렬을 입력하세요.: ')
  num2 = input('두 번째 정렬을 입력하세요.: ')

  print(isPalindrome(num1, num2))
```

15 역순 연결 리스트

```python
from typing import Sequence

def isPalindrome(a: Sequence) -> str:

  new = []
  for i in a:
    if i.isdigit():
      new.append(i)

  new.reverse()
  new = '->'.join(new)

  return new

if __name__ == '__main__':
  num = input('정렬을 입력하세요.: ')

  print(isPalindrome(num))  
```
